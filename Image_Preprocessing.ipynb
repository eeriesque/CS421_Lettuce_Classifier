```python
# ls_bacterial
from numpy import expand_dims
from keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rotation_range = 90,
    horizontal_flip = True,
    height_shift_range = 0.5,
    brightness_range = [0.2, 1.0],
    zoom_range = [0.5, 1.0]
)
```

```python
start = time.time()
for i in ls_bacterial:
    f, e = os.path.splitext(i)
    file_name = os.path.basename(f)
    im_resized
    if os.path.isfile(i):
        im = PIL.Image.open(i)
        im_resized = im.resize(img_size, PIL.Image.ANTIALIAS)
#         im_resized.save(f + '_resized.jpg', 'JPEG', quality = 90)
    
#     img = load_img(i)
    data = img_to_array(im_resized)
    sample = expand_dims(data, 0)
    
    it = datagen.flow(sample, batch_size = batch_size)
    
    for i in range(16):
#         plt.subplot(4, 4, i + 1)
        
        # generate batch of images
        batch = it.next()
        
        # convert to unsigned integers for viewing
        image = batch[0].astype('uint8')
#         print(image.dtype)
        
        # plot raw pixel data
        plt.imshow(image)
    
        img = PIL.Image.fromarray(image, mode = 'RGB')
        new_path = bacterial_augmented_path + '\\' + file_name + '_processed-' + str(i)
#         plt.savefig(new_path + '.jpg')
    
        img.save(new_path + '.jpg', 'JPEG')
        
#     plt.show()
print('Time Lapsed:\t', str(round(time.time() - start, 4)))
break
```
